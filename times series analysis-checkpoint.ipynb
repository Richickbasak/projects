```python
import pandas as pd
import numpy as np
import warnings

# Create a date range for 5 years of monthly data
dates = pd.date_range(start='2018-01-01', end='2022-12-01', freq='MS')

# Create synthetic sales data with trend + noise (no seasonality)
np.random.seed(42)
n = len(dates)
trend = np.linspace(100, 500, n)               # upward trend
noise = np.random.normal(loc=0, scale=20, size=n)  # random noise
sales = trend + noise


# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Sales': sales})
df.set_index('Date', inplace=True)

# Display head of dataset
print(df.head())
```

# stationarity 

```python
import statsmodels.tsa.stattools as sts
```

```python
result = sts.adfuller(df.Sales)
```

```python
print(f'Test Statistic : {result[0]:.4f}')
print(f'  p-value        : {result[1]:.4f}')
print(f'  # Lags Used    : {result[2]}')
print(f'  # Observations : {result[3]}')
print(f'  Critical Values: {result[4]}')
if result[1] < 0.05:
     print('  → Stationary')
else:
    print('  → Non-stationary')
```

# making the series stationary

```python
# differencing

df['Sales'] = df['Sales'].diff()
df.dropna(inplace = True)
df.head()
```

```python
result = sts.adfuller(df.Sales)

print(f'Test Statistic : {result[0]:.4f}')
print(f'  p-value        : {result[1]:.4f}')
print(f'  # Lags Used    : {result[2]}')
print(f'  # Observations : {result[3]}')
print(f'  Critical Values: {result[4]}')
if result[1] < 0.05:
     print('  → Stationary')
else:
    print('  → Non-stationary')
```

```python
df.Sales.plot(figsize = (20,5), title = 'first order differenced series')
plt.show()
```

# AR model using sarimax

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_ar = SARIMAX(df_train.Sales,order=(1,0,0),seasonal_order=(0,0,0,0))
results_ar = model_ar.fit(disp=False)
print(results_ar.summary())
```

```python
forecaste = results_ar.get_forecast(steps=len(df_test))
pred = forecaste.predicted_mean
conf_int = forecaste.conf_int()

# plot
plt.figure(figsize=(20,5))
plt.plot(df_train,label='Train')
plt.plot(df_test.index,df_test,label='test',color='red')
plt.plot(df_test.index,pred,label='forcaste',color='orange')
plt.fill_between(df_test.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='orange', alpha=0.3)
plt.title("SARIMAX Forecast on Test Data")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(df_test, pred)
mae = mean_absolute_error(df_test, pred)

print(f'MSE: {mse:.2f}, MAE: {mae:.2f}')
```

```python
warnings.filterwarnings("ignore")

best_aic = float("inf")
best_p = 0

for p in range(1, 11):  # Trying AR orders from 1 to 10
    try:
        model = SARIMAX(df_train['Sales'], order=(p, 0, 0), seasonal_order=(0, 0, 0, 0))
        result = model.fit(disp=False)
        if result.aic < best_aic:
            best_aic = result.aic
            best_p = p
    except Exception as e:
        print(f"Error for p={p}: {e}")
        continue

print(f"Best AR order: p={best_p}, AIC={best_aic:.2f}")
```

# AUTO ARIMA

```python
from pmdarima.arima import auto_arima
```

```python
# Auto ARIMA model
model = auto_arima(df_train['Sales'],
                   seasonal=True,      # Set to False if data is non-seasonal
                   m=12,               # Frequency of seasonality (e.g. 12 for monthly)
                   trace=True,         # Prints progress
                   error_action='ignore',  
                   suppress_warnings=True,
                   stepwise=True)

# Summary
print(model.summary())
```

```python
# Predict the next N periods (same length as test set)
n_periods = len(df_test)
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Convert to pandas series with test index
forecast_series = pd.Series(forecast, index=df_test.index)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_train['Sales'], label='Train')
plt.plot(df_test['Sales'], label='Test')
plt.plot(forecast_series, label='Predicted')
plt.fill_between(df_test.index, conf_int[:, 0], conf_int[:, 1], color='lightgray', alpha=0.5)
plt.legend()
plt.title('Auto ARIMA Forecast vs Actual')
plt.show()
```

```python
# Forecast future periods beyond test data
future_periods = 24
future_forecast, future_conf_int = model.predict(n_periods=future_periods, return_conf_int=True)

future_index = pd.date_range(df_test.index[-1] + pd.offsets.MonthBegin(), periods=future_periods, freq='M')
future_series = pd.Series(future_forecast, index=future_index)

plt.figure(figsize=(10, 5))
plt.plot(df['Sales'], label='All Data')
plt.plot(future_series, label='Future Forecast')
plt.fill_between(future_index, future_conf_int[:, 0], future_conf_int[:, 1], color='lightblue', alpha=0.5)
plt.legend()
plt.title('Future Forecast with Auto ARIMA')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

rmse = mean_squared_error(df_test['Sales'], forecast_series, squared=False)
mae = mean_absolute_error(df_test['Sales'], forecast_series)

print(f'RMSE: {rmse:.2f}')
print(f'MAE : {mae:.2f}')
```
