# removing stop words

```python
from nltk.corpus import stopwords

# Load English stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(text):
    words = text.split()  # Split text into words
    return ' '.join([word for word in words if word.lower() not in stop_words])

# Apply the function to the 'text' column
df['reviewDescription'] = df['reviewDescription'].apply(remove_stopwords)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_sentence(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)
    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    # Join the lemmatized words back into a sentence
    return ' '.join(lemmatized_words)

# Apply the lemmatization process to each sentence in each review
df["reviewDescription"] = df["reviewDescription"].apply(
    lambda review: [lemmatize_sentence(sentence) for sentence in sent_tokenize(review)] if isinstance(review, str) else review
)
df["variant"] = df["variant"].apply(
    lambda review: [lemmatize_sentence(sentence) for sentence in sent_tokenize(review)] if isinstance(review, str) else review
)
```

# Removing Emojis

```python
import re

# Function to remove emojis
def remove_emojis(text):
    # This pattern will match most emojis
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F1E0-\U0001F1FF"  # Flags (iOS)
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed characters
        "]+", flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Apply to DataFrame column
df["reviewDescription"] = df["reviewDescription"].apply(lambda x: remove_emojis(x) if isinstance(x, str) else x)
df["reviewDescription"]
```

# Removing Punctuations

```python
from string import punctuation
```

```python
def remove_punctuation(text):
    # Convert to string if it's not already a string
    text = str(text)
    tokens = word_tokenize(text)  # Tokenize the text
    clean_tokens = [word for word in tokens if word not in punctuation]  # Remove punctuation
    return ' '.join(clean_tokens)  # Join the tokens back into a clean sentence

# Apply the function to each row in the 'reviewDescription' column
df["reviewDescription"] = df["reviewDescription"].astype(str).apply(remove_punctuation)

df["variant"] = df["variant"].astype(str).apply(remove_punctuation)
```

# Sentiment Analysis

```python
from textblob import TextBlob
```

```python
def get_polarity(review):
    return TextBlob(review).sentiment.polarity
```

```python
def analysis(score):
    if score < 0 :
        return 'negative'
    elif score > 0:
        return 'positive'
    else:
        return "neutral"
```

```python
df["reviewDescription_polarity"] = df["reviewDescription"].apply(get_polarity)
df['reviewDescription_polarity_score'] = df["reviewDescription_polarity"].apply(analysis)
```

```python
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def vas(review):
    vas = analyzer.polarity_scores(review)
    return vas['compound']

def analysis_vas(score):
    if score <=-0.5:
        return 'negative'
    elif score >= 0.5:
        return 'positive'
    else:
        return "neutral"
```

```python
df['reviewDescription_vader_sentiment'] = df["reviewDescription"].apply(vas)
df['reviewDescription_vader_sentiment_score'] = df["reviewDescription_vader_sentiment"].apply(analysis_vas)
df
```

```python
# Define a regex pattern to capture size and service provider
pattern = r"size (\d{3,4} ?gb)(?:.*service provider (\w+))?"

# Apply the regex to extract the details into new columns
df['size'] = df['variant'].apply(lambda x: re.search(pattern, x).group(1) if re.search(pattern, x) else None)
df['service_provider'] = df['variant'].apply(lambda x: re.search(pattern, x).group(2) if re.search(pattern, x) else 'unlocked')
df
```

```python
print(df["size"].unique())
```

```python
# Define a function to add a space between the number and 'gb'
def format_size(size):
    return re.sub(r"(\d+)(gb)", r"\1 gb", size)

# Apply the function to standardize the size column
df['size'] = df['size'].apply(format_size)

print(df["size"].unique())
```
