# lets us consider there are 7 stocks in the portfolio for the analysis

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'NVDA', 'TSLA']  # example tech tickers on Yahoo
start = "2022-01-01"
end = "2025-11-21"
```

```python
prices = yf.download(tickers, start=start, end=end, auto_adjust=False)["Adj Close"].dropna()
prices
```

```python
daily_returns = prices.pct_change().dropna()
daily_returns
```

```python
trading_days = 252
```

```python
mu = daily_returns.mean() * trading_days            # expected annual returns
mu
```

```python
sigma = daily_returns.cov() * trading_days          # annual covariance
sigma
```

```python
rf = 0.0022   # risk free rate
```

```python
corr = daily_returns.corr()
corr
```

```python
n = len(tickers)

def portfolio_stats(weights,mu,sigma):
    ret = weights.dot(mu)
    vol = np.sqrt(weights.T @sigma @weights)
    Sharpe = (ret - rf)/ vol
    return ret, vol, Sharpe

cons = ({'type':'eq','fun': lambda w: np.sum(w)-1}) 

bounds = tuple((0,0.3) for _ in range(n))  

init = np.repeat(1/n, n)

min_var = minimize(lambda w: w.T @ sigma @ w,
                   x0=init,
                   method='SLSQP',
                   bounds=bounds,
                   constraints=cons)

w_minvar = min_var.x
ret_minvar, vol_minvar, _ = portfolio_stats(w_minvar, mu, sigma)
```

```python
# Max Sharpe (tangency) portfolio
def neg_sharpe(w, mu, sigma, rf=rf):
    ret = w.dot(mu)
    vol = np.sqrt(w.T @ sigma @ w)
    return - (ret - rf) / vol

max_sharpe = minimize(neg_sharpe, x0=init, args=(mu,sigma),
                      method='SLSQP', bounds=bounds, constraints=cons)

w_maxsharpe = max_sharpe.x
ret_maxsharpe, vol_maxsharpe, sharpe_max = portfolio_stats(w_maxsharpe, mu, sigma)
```

```python
# Efficient frontier (quadratic-program approach or Monte-Carlo)
# simple Monte-Carlo approximation:
n_port = 40000
weights = np.random.dirichlet(np.ones(n), n_port)
rets = weights @ mu.values
vols = np.sqrt(np.einsum('ij,jk,ik->i', weights, sigma.values, weights))

# Plot
plt.figure(figsize=(10,6))
plt.scatter(vols, rets, s=2, alpha=0.25)
plt.scatter([vol_minvar],[ret_minvar], marker='*', s=200, label='Min Variance')
plt.scatter([vol_maxsharpe],[ret_maxsharpe], marker='D', s=120, label='Max Sharpe')
plt.xlabel('Annual Volatility')
plt.ylabel('Annual Return')
plt.title('Efficient Frontier (approx)')
plt.legend()
plt.show()
```

```python
#Results table
results = pd.DataFrame({
    'Ticker': tickers,
    'MinVar_w': np.round(w_minvar,4),
    'MaxSharpe_w': np.round(w_maxsharpe,4)
})
print(results)
print("MinVar: Return %.3f Vol %.3f" % (ret_minvar, vol_minvar))
print("MaxSharpe: Return %.3f Vol %.3f Sharpe %.3f" % (ret_maxsharpe, vol_maxsharpe, sharpe_max))
```

This portfolio focuses purely on reducing risk, not boosting returns.

It avoids volatile names like NVDA and TSLA entirely.

It concentrates more on stable tech giants: AAPL, GOOGL, META.

Volatility is low-ish (0.258), but return is also moderate (0.168).

This is expected — min variance always prefers low covariance + low volatility assets, even if returns are not very high.

This portfolio wants maximum return per unit risk, so it shifts heavily toward:

NVDA (0.30)

GOOGL (0.30)

AAPL (0.26)

These are high-return, high-quality growth stocks.

MSFT and TSLA get 0% weight because their return-to-risk ratio (Sharpe contribution) is not competitive.

Vol is higher (0.324), but return jumps to 0.322, producing the best Sharpe ≈ 1.

```python
# Print the individual Sharpe ratios
for ticker in tickers:
    individual_sharpe = (mu[ticker] - rf) / sigma.loc[ticker, ticker]**0.5
    print(f"{ticker}: {individual_sharpe:.4f}")
```
